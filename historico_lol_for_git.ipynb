```python
import pandas as pd
```

```python
#Baixando o CSV com os dados de 2011-2022

df = pd.read_csv('champions_stats.csv')

df.head()
```

```python
#FaÃ§o um tratamento para conciliar Season com o ano vigente 
#Como a ideia Ã© virar um Index, Ã© mais tranquilo organizar por ano do MSI/Mundial
 
df['ano'] = pd.to_datetime(2000 + 10 + df['season'], format='%Y') + pd.offsets.DateOffset(months=0, days=0)
df['ano'] = pd.to_datetime(df['ano']).dt.year
df.head()
```

```python
df.head()
```

```python
#Criando o dataframe pro grÃ¡fico

df_new = df[['champion', 'games_contests', 'ano']]
df_new.head()
```

```python
#Nesta etapa o que fazemos Ã© somar ano a ano os campeÃµes para criar um acumulado.

df_chart = df_new.pivot_table(
    index='ano',
    columns='champion',
    values='games_contests',
    aggfunc='sum'  # ou 'mean', 'max', etc. conforme seu caso
)

df_chart = df_chart.rename_axis(None, axis=1).reset_index()
```

```python
#Conferindo o resultado
df_chart.head()
```

```python
#Setando o ano como Index
df_chart.set_index('ano', inplace=True)

#Trocando NaN por 0 
df_chart = df_chart.fillna(0)
```

```python
#O cumsum faz o acumulado dos dados.
df_chart_acumulado = df_chart.cumsum()
```
