```python
df_new = df[['champion', 'games_contests', 'season']]
df_new.head()
```

```python
df_chart = df_new.pivot_table(
    index='season',
    columns='champion',
    values='games_contests',
    aggfunc='sum'  # ou 'mean', 'max', etc. conforme seu caso
)

df_chart = df_chart.rename_axis(None, axis=1).reset_index()
```

```python
df_chart.head()
```

```python
df_chart.set_index('season', inplace=True)

# Opcional: preencher NaNs com 0 (ou use outra lÃ³gica)
df_chart = df_chart.fillna(0)
```

```python
df_chart_acumulado = df_chart.cumsum()
```
